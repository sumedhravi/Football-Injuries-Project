```python
def prepare_injury(injury):
    injury=injury[injury['team']!='fc-fulham']
    injury=injury[injury['team']!='sheffield-united']
    injury=injury[injury['team']!='west-bromwich-albion']
    injury=injury[injury['team']!='sd-eibar']
    injury=injury[injury['team']!='real-valladolid']
    injury=injury[injury['team']!='sd-huesca']
    injury=injury[injury['team']!='ac-florenz']
    injury=injury[injury['team']!='parma-calcio-1913']
    injury=injury[injury['team']!='benevento-calcio']
    injury=injury[injury['team']!='fc-crotone']
    injury=injury[injury['team']!='fc-schalke-04']
    injury=injury[injury['team']!='sv-werder-bremen']
    injury['team'][injury['team']=='athletic-bilbao']='Athletic Club'
    injury['team'][injury['team']=='real-betis-sevilla']='Betis'
    injury['team'][injury['team']=='deportivo-alaves']='Alavés'
    injury['team'][injury['team']=='fc-cadiz']='Cádiz'
    injury['team'][injury['team']=='ssc-neapel']='Napoli'
    injury['team'][injury['team']=='ac-mailand']='Milan'
    injury['team'][injury['team']=='as-rom']='Roma'
    injury['team'][injury['team']=='fc-turin']='Torino'
    injury['team'][injury['team']=='genua-cfc']='Genoa'
    injury['team'][injury['team']=='rasenballsport-leipzig']='RB Leipzig'
    injury['team'][injury['team']=='borussia-monchengladbach']="M'Gladbach"
    injury['team'][injury['team']=='1-fc-koln']='Köln'
    
    return injury
```
