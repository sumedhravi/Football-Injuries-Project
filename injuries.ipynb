```python
# Imports
import numpy as np
import pandas as pd
url = "https://www.kaggle.com/eliesemmel/soccerplayersinjuries"
nbk = "https://www.kaggle.com/eliesemmel/kernelde4fae5abe/notebook"
```

```python
data = []
injuries = []
transfers = []
i=0
for t in open('Final-player.txt'):
    i=i+1
    cur = None
    if i == 2580:
        cur = eval(t[1:-1])
    else:
        cur = eval(t[1:-2])
    data.append(cur)
    for injury in cur[-1]:
        injuries.append([cur[0], cur[1], injury[0], injury[1], injury[2]])
    for transfer in cur[-2]:
        transfers.append([cur[0], cur[1], transfer[0], transfer[1], transfer[2], transfer[3]])
df = pd.DataFrame(data)
# coercing to types and cleaning data
df.columns = ["id", "name", "club","club_value","birth","weight","height","country","role","foot","transfers","injuries"]
df['id'] = df['id'].astype(int)
df['name'] = df['name'].astype(str)
df['club'] = df['club'].astype(str)
df["club_value"] = df["club_value"].astype(str).apply(lambda x:
    int(float(x.split(" ")[0].replace(",", "."))*(1e9 if x.split(" ")[1] == "Bill." else 1e6)))
df['birth'] = pd.to_datetime(df['birth'], errors='coerce')
df.loc[df.weight == '-', "weight"] = None
df['weight'] = df['weight'].astype('float').astype("Int64")
df['height'] = pd.to_numeric(df['height'].str.strip().str.replace(",", "."), errors='coerce')
df['role'] = df['role'].astype(str)
df[["Line", "Wing", "Position"]] = df["role"].str.replace("-", " ").str.split("\s+", expand=True)
df['Line'] = df['Line'].astype(str).str.strip()
df['Wing'] = df['Wing'].astype(str).str.strip()
df['Position'] = df['Position'].astype(str).str.strip()
df['foot'] = df['foot'].astype(str)
df.loc[~((df.foot == "left") | (df.foot == "right") | (df.foot == "both") | (df.foot == "None")), ["foot"]] = "None"
number_of_players = df.shape[0]
print(df.columns)
df.dtypes
```

```python
df
```

```python
df_injuries = pd.DataFrame(injuries, columns=["id", "name", "season", "description", "days"])
df_injuries['id'] = df_injuries['id'].astype(int)
df_injuries['name'] = df_injuries['name'].astype(str)
df_injuries['season'] = df_injuries['season'].astype(str)
df_injuries['description'] = df_injuries['description'].astype(str)
df_injuries['days'] = df_injuries['days'].astype(int)
df_injuries[['start_year','end_year']] = df_injuries['season'].str.split('/', expand=True)
df_injuries['start_year'] = df_injuries['start_year'].astype(int).apply(lambda x: x + 2000 if x <= 22 else x + 1900)
df_injuries['end_year'] = df_injuries['end_year'].astype(int).apply(lambda x: x + 2000 if x <= 22 else x + 1900)
df_injuries
```

```python
df_transfers = pd.DataFrame(transfers, columns=["id", "name", "club_from", "club_to", "season", "date"])
df_transfers['id'] = df_transfers['id'].astype(int)
df_transfers['name'] = df_transfers['name'].astype(str)
df_transfers['club_from'] = df_transfers['club_from'].astype(str).str.strip()
df_transfers['club_to'] = df_transfers['club_to'].astype(str).str.strip()
df_transfers['season'] = df_transfers['season'].astype(str)
df_transfers[['start_year','end_year']] = df_transfers['season'].str.split('/', expand=True)
df_transfers['start_year'] = df_transfers['start_year'].astype(int).apply(lambda x: x + 2000 if x <= 22 else x + 1900)
df_transfers['end_year'] = df_transfers['end_year'].astype(int).apply(lambda x: x + 2000 if x <= 22 else x + 1900)
df_transfers["date"] = pd.to_datetime(df_transfers["date"], format="%b %d, %Y")
df_transfers
```

```python
injury_per_player = df.merge(df_injuries, on=["id", "name"])
injury_per_player
```

```python
# Saving generated data frames
df.to_pickle("./Injury/cleaned_df.pkl")
df_injuries.to_pickle("./Injury/cleaned_injuries.pkl")
df_transfers.to_pickle("./Injury/cleaned_transfers.pkl")
```
